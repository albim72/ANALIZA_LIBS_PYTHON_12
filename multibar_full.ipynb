```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler                                                               
```

```python
osoby = ("Anna","Piotr","Olga","Roman")
informacje = {
    'wiek':(34,56,23,45),
    'waga':(56.8,89.6,60.3,103.6),
    'wzrost':(162,172,170,189)
}
```

```python
x = np.arange(len(osoby))
width = 0.25
multiplier = 0

fig,ax = plt.subplots(constrained_layout=True)
for attribute,mesurement in informacje.items():
    offset = width*multiplier
    rects = ax.bar(x+offset,mesurement,width,label=attribute)
    ax.bar_label(rects,padding=5)
    multiplier += 1

ax.set_ylabel('cecha')
ax.set_title("wybrane informacje na temat os√≥b...")
ax.set_xticks(x+width,osoby)
ax.legend(loc='upper left',ncols=3)
ax.set_ylim(0,200)
plt.show()
```

```python
#przygotowanie danych
cases = [
    None,
    8,
    (30,8),
    [16,24,30],
    [0,-1],
    slice(200,200,3),
    0.1,
    0.3,
    (0.0,0.1),
    (0.45,0.1)
]

colors =[
    '#1f77b4',
    '#ff7f0e',
    '#2ca02c',
    '#d62728',
    '#9467bd',
    '#8c564b',
    '#e377c2',
    '#7f7f7f',
    '#bcbd22',
    '#17becf',
]
```

```python
mpl.rcParams['axes.prop_cycle'] = cycler(markevery=cases,color=colors)
x = np.linspace(0,2*np.pi)
offsets = np.linspace(0,2*np.pi,11,endpoint=False)
yy = np.transpose([np.sin(x+phi) for phi in offsets])

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.6,0.75])

for i in range(len(cases)):
    ax.plot(yy[:,i],marker='o',label=str(cases[i]))
    ax.legend(bbox_to_anchor =(1.05,1), loc='upper left', borderaxespad=0.)

plt.title('wy≈õwitlanie grupy wykres√≥w cyklicznych')
plt.show()
```
